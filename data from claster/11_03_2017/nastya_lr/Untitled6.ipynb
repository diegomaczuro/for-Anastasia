```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse
%matplotlib inline

NUM = 250

ells = [Ellipse(xy=rnd.rand(2)*10, width=rnd.rand(), height=rnd.rand(), angle=rnd.rand()*360)
        for i in range(NUM)]

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```

```python
tenzor_array = []
for i in tqdm(xrange(len(e1))):
    ta1 = []
    for j in xrange(len(e1[0])):
        ta2 = []
        for k in xrange(len(e1[0][0])):
            D = np.array( [[b11[i][j][k], b21[i][j][k], b31[i][j][k]],
                          [b12[i][j][k], b22[i][j][k], b32[i][j][k]],
                          [b13[i][j][k], b23[i][j][k], b33[i][j][k]]])
            E = np.array([[e1[i][j][k], 0, 0],
                         [0, e2[i][j][k], 0],
                         [0, 0, e3[i][j][k]]])
            t_ = np.dot(D, E)
            t = np.dot(t_, D.T)
            ta2.append(t)
            
        ta1.append(ta2)
    
    tenzor_array.append(ta1)
```
